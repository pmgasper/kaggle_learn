## Level 4: Grouping and Sorting

```python
import matplotlib.pylab as plt
import pandas as pd

pd.set_option('max_rows', 5)
```

```python
reviews = pd.read_csv("data/winemag-data-130k-v2.csv", index_col=0)
reviews
```

### Grouping

groupby
  
    Groups a series using a mapper
    Returns DataFrameGroupBy object
      Contains a DataFrame for each group
   
DataFrameGroupBy object methods

    .apply
       apply a function to each the dataframe of each group
       returns applied series or dataframe

    .agg
       aggregates each group by specified function(s)
       returns DataFrame for aggregates
  

multi-indexes

     groupby on multiple columns returns dataframes with two (or more) levels 
     of labels required to retrieve a value
  
     .reset_index()
       moves index levels to new columns and creates a new sequential index 

```python
# groupby points
# groupby - returns DataFrameGroupBy object
# DataFrameGroupBy.column - returns SeriesGroupBy object
# SeriesGroupBy.count - return count each series group as a series
print(reviews.groupby('points'))
print(reviews.groupby('points').points)
print()
print(reviews.groupby('points').points.count())
print(type(reviews.groupby('points').points.count()))
```

```python
# First title for each winery
print(reviews.groupby('winery').apply(lambda df: df.title.iloc[0]))
print()

# Tasters by winery
print(reviews.groupby('winery').apply(lambda df: (df.taster_name.unique())))
print()

# Shortest wine title by winery 
print(reviews.groupby('winery').apply(lambda df: 
                                      df.title.loc[df.title.str.len().idxmin()]))

# Highest points for each country - provience pair
reviews.groupby(['country', 'province']).apply(lambda df: df.loc[df.points.idxmax()])
```

```python
# Highest and lowest by variery - .agg
wine_price_extremes = reviews.groupby('variety').price.agg(['min', 'max'])
print(type(wine_price_extremes))
wine_price_extremes
```

```python
# Multiple column aggregate
wine_price_extremes = reviews.groupby('variety').agg({'price':['min', 'max'],
                                                'points':['sum', 'mean', 'count']})
wine_price_extremes
```

### Sorting

df.sort_values(by)

df.sort_index

```python
# Sort by values
reviews.sort_values(by='points', ascending=False)
```

```python
# Sort by values, include secondary sort column 
reviews.sort_values(by=['points', 'price'], ascending=False)
```

```python
# grouped by column / sorted by count of non-nan enteries in that column
common_wine_reviewers = reviews.groupby(['taster_twitter_handle']).\
                        taster_twitter_handle.count().sort_values(ascending=False)
print(common_wine_reviewers)
```

```python
# Highest points per price
best_wine = reviews.groupby(['price']).points.max().sort_index()
print(best_wine)
print()

# Also show corresponding title
best_wine = reviews.loc[reviews.groupby(['price']).points.idxmax()]
best_wine.loc[:, ['price', 'points', 'title']].sort_values('price')
```

```python
# Group means
reviewer_mean_ratings = reviews.groupby(['taster_name']).points.mean().\
                                                         sort_values(ascending=False)
print(reviewer_mean_ratings)
reviewer_mean_ratings.plot.bar()
plt.show()
```

```python
# Group mins and maxes
wine_price_range = reviews.groupby(['variety']).price.agg(['min', 'max']).\
                                   sort_values(['min', 'max'], ascending=False)
display(wine_price_range)                                                          
wine_price_range.head().plot.bar() 
plt.show()
```

```python
# Count multi-index pairs
reviews['n'] = 0
country_variety_pairs = reviews.groupby(['country', 'variety']).n.count().\
                                         sort_values(ascending=False)
display(country_variety_pairs)
country_variety_pairs.head(10).plot.bar()

plt.show()
```
