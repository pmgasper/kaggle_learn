### Part 2: Univariate plotting with pandas

Panda's single-variable visualizations:

**df.plot.bar()** and **df.plot.barh()**  <br> 
* Good for nominal and small ordinal categorical data.<br>
* Height can represent anything, as long as it is a number<br>
* Each bar can represent anything, as long as it is a category.<br>
* Start getting difficult to read around 20 categories<br> 
<br>
    
**df.plot.pie()**<br>
* Bar charts drawn in a circle<br>
* Not as good as a bar chart for analysis<br>
* Difficult to compare slices of similar size<br>
<br>

**df.plot.line()**
* Good for ordinal categorical and interval data.<br>
* Especially good for data with many unique values.<br>
* It can be harder to distinguish individual values than on a bar chart.<br>
* The continuous nature makes them inappropriate for nominal data.<br>
  However, nominal data can be converted into percentiles and ordered<br>
  (i.e. city streets ordered by the number of restaurants on that street)<br>
<br>

**df.plot.area()**
* Line charts with the bottom shaded in.<br>
* Good for the same things, ordinal categorical and interval data.<br>
<br>

**df.plot.hist()**
* Good for interval data<br>
* Bar charts where the categories are bins (ranges of data)<br>
* Even sized bins don't deal very well with skewed data<br>
  One way to deal with skew is restrict the range graphed to not include extremes<br>
  Other methods exist, but are not discusses in this tutorial<br>

```python
# Dataframe for examples
reviews = pd.read_csv("../pandas/data/winemag-data_first150k.csv", index_col=0)
reviews
```

```python
# Bar chart - count
reviews['province'].value_counts().head(10).plot.bar()
plt.show()
```

```python
# Horizontal bar chart - count
reviews['province'].value_counts().head(10).plot.barh()
plt.gca().invert_yaxis() # Plot top to bottom
plt.show()
```

```python
# Pie chart
reviews['province'].value_counts().head(10).plot.pie()

# Ensure pie is a circle, not oval.
plt.gca().set_aspect('equal')
plt.show()
```

```python
# Bar chart - proportions 
((reviews['province'].value_counts().head(10) / len(reviews))
 .plot.bar())
plt.show()
```

```python
# Bar chart - 20 categories
reviews['points'].value_counts().sort_index().plot.bar(color='steelblue')
plt.show()
```

```python
# Line chart
reviews['points'].value_counts().sort_index().plot.line()
plt.show()
```

```python
# Area Chart
reviews['points'].value_counts().sort_index().plot.area()
plt.show()
```

```python
# Histogram
reviews[reviews['price'] < 200]['price'].plot.hist()
plt.show()
```

```python
reviews['points'].plot.hist()
plt.show()
```
