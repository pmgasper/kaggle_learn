### Bivariate plotting with Pandas

Panda's two-variable visualizations:

**df.plot.scatter()**
* Good for interval and some nominal categorical data<br>
* Maps each variable to a point in two-dimensional space<br>
* Naive scatter plots do not effectively treat points which map to the same place.<br>
<br>
    
**df.plot.hexbin()**
* Good for interval and some nominal categorical data.<br>
* Aggregates points in space into hexagons, and then colorize those hexagons<br>
<br>

**df.plot.bar(stacked=True)**
* Good for nominal and ordinal categorical data.<br>
* Counts are colored by column category and stacked at corresponding row indexes across the x<br>
* The number of stacked categories is limited (generally < 8) before the plot becomes difficult to interpret.<br>
* Comparison of categories can also be difficult<br>
<br>

**df.plot.area()**
* Overlapping area line plots<br>
* Same weaknesses as stacked bar<br>
<br>

**df.plot.line()**
* Good for ordinal categorical and interval data.<br>
* More interpretable than stacked plots

```python
# Dataframe for examples
reviews = pd.read_csv("../pandas/data/winemag-data_first150k.csv", index_col=0)
reviews
```

```python
# Scatter plot
  # price < 100 cuts off a few outliers to better disperse the points of interest
  # sample(100) randomly selects 100 points to minimize overlapping points
reviews[reviews['price'] < 100].sample(100).plot.scatter(x='price', y='points')
plt.show()
```

```python
# Scatter plot - all data with2 price < 100 
reviews[reviews['price'] < 100].plot.scatter(x='price', y='points')
plt.show()
```

```python
# Hex plot
reviews[reviews['price'] < 100].plot.hexbin(x='price', y='points', gridsize=15)
plt.show()
```

```python
# DataFrame for examples
#   Contingency table - expected format for many Pandas multivariate plots
wine_counts = pd.read_csv("data/top-five-wine-score-counts.csv", index_col=0)
wine_counts
```

```python
# Stacked Bar
wine_counts.plot.bar(stacked=True)
plt.show()
```

```python
# Area
wine_counts.plot.area()
plt.show()
```

```python
# Bivariate line 
wine_counts.plot.line()
plt.show()
```

#### Exercises

```python
pd.set_option('max_columns', None)
pokemon = pd.read_csv("data/pokemon.csv")
pokemon.head(3)
```

```python
# Scatter
pokemon.plot.scatter(x='attack', y='defense')
plt.show()
```

```python
# Hex
pokemon.plot.hexbin(x='attack', y='defense', gridsize=15)
plt.show()
```

```python
pokemon_stats_legendary = (pokemon.groupby(['is_legendary', 'generation'])
                           .mean()[['attack', 'defense']])

pokemon_stats_legendary.plot.bar(stacked=True)
plt.show()
```

```python
pokemon_stats_by_generation = (pokemon.groupby('generation').mean()
                               [['hp', 'attack', 'defense', 
                                 'sp_attack', 'sp_defense', 'speed']])

pokemon_stats_by_generation.plot.line()
plt.show()
```
