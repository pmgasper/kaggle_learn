## Level 1, Part 4 - Styling your plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
#plt.style.available
pd.set_option('max_rows', 9)
```

### Styling your plots



```python
# Dataframe for examples
reviews = pd.read_csv("../pandas/data/winemag-data_first150k.csv", index_col=0)
reviews.head(3)
```

```python
# Pandas - bar plot - default styling
plt.style.use('default')
reviews['points'].value_counts().sort_index().plot.bar()
plt.show()
```

```python
# Pandas - bar plot - with styling

# If usng a Style class, it must be set before plotting
plt.style.use('fivethirtyeight')

# Creating a matplotlib figure instance
fig = plt.figure(figsize=(12,6))
print(type(fig))

# figsize = (width, height) in inches
  # Figure size can also be adjusted after creation
    # fig.set_size_inches(8, 6)
  # In scatter plot in exercises neither of these methods seems to work
    # and figure size must be set using a keyword when plotting

# Plotting with Pandas
#   Assigning plot to axis instance for additional styling options
#   Some styling is easiest to do on plot creation
#     color = colors for bars, or list of colors that will be iterated
#     fontsize = set fontsize to text included in initial plot
ax = reviews['points'].value_counts().sort_index().plot.bar(
            color='mediumvioletred',
            fontsize=16)
        
# Add title with custom fontsize           
ax.set_title("Rankings Given by Wine Magazine", fontsize=24)

# Add x and y axis labels
ax.set_xlabel('Points', fontsize=20)
ax.set_ylabel('Count', fontsize=20)

# Get axis limits
  # x_min, x_max, y_min, y_max
  # note x range for bar plot was set for the number of catagories 
print(plt.axis())

# Set axis limits
  # defaults do well here, but could be adjusted with xlim or ylim
# plt.xlim(-0.5, 20.5)
# plt.ylim(0, 22000)

# Remove border
#  This is already done by the 'fivethirtyeight' style, but can also be done
#  using Seaborn despine
sns.despine(bottom=True, left=True)

plt.show()
```

```python
# Seaborn - count plot - with styling

# If usng a Style class, it must be set before plotting
plt.style.use('fivethirtyeight')

# Creating a matplotlib figure instance
fig = plt.figure(figsize=(12,6))
print(type(fig))

# figsize = (width, height) in inches
  # Figure size can also be adjusted after creation
  # fig.set_size_inches(8, 6)

# Still figuring out fontsize in Seaborn
#   setting font_scale seems to affect all default font sizes in all subsequent 
#   Seaborn plots
sns.set(font_scale=1.5)

# Plotting with Seaborn
#   Assigning plot to axis instance for additional styling options
ax = sns.countplot(reviews['points'])
 
    
# Add title with custom fontsize           
ax.set_title("Rankings Given by Wine Magazine", fontsize=24)

# Add x and y axis labels
ax.set_xlabel('Points', fontsize=20)
ax.set_ylabel('Count', fontsize=20)

# Get axis limits
  # x_min, x_max, y_min, y_max
  # note x range for bar plot was set for the number of catagories 
# print(plt.axis())

# Set axis limits
  # defaults do well here, but could be adjusted with xlim or ylim
# plt.xlim(-0.5, 20.5)
# plt.ylim(0, 22000)

# Remove border
#  This is already done by the 'fivethirtyeight' style, but can also be done
#  using Seaborn despine
sns.despine(bottom=True, left=True)

plt.show()

# Resetting font scale
sns.set(font_scale=1)
```

```python
# Seaborn - countplot - no styling (besides default)

sns.set(font_scale=1)

sns.countplot(reviews['points'])
plt.show()
```
